# Bayesian model

Bayes' rule 

$P(\theta|y)=\frac{P(\theta) P(y|\theta)} {P(y)}$


We can write out the following probabilistic model (now explicitly specifying probability distributions):

- $\mu$∼Normal(0,1)
- $\sigma$∼Gamma(11,1)
- $y_{obs}$∼Normal($\mu$,$\sigma$)

Let's now map the symbols onto Bayes' rule.

- $\theta$ are the parameters, which are $\mu$ and $\sigma$ here. 
- $y_{obs}$ is the data that I will observe
- $P(\theta|y_{obs})$ is the posterior, which we would like to compute.
- $P(y|\theta)$ is the likelihood, and is given by y's probability distribution Normal($\mu$,$\sigma$), or in probability notation, $P(y|\mu,\sigma)$.
- $P(\theta)$ is the the prior, and is given by $P(\mu,\sigma)$.
- $P(y_{obs})$ is a hard quantity to calculate, so we sort of cheat and don't use it, and merely claim that the posterior is proportional to likelihood times prior.

```python
class Normal:
    
    """
    Normal distribution 
    """
    def __init__(self, mu, sigma):
        self.mu = mu
        self.sigma = sigma
        self.dist = norm(loc=mu, scale=sigma)

    def pdf(self, x):
        return self.dist.pdf(x)

    def logpdf(self, x):
        return self.dist.logpdf(x)
    
    def sample(self, n):
        return self.dist.rvs(n)

class Gamma:
    
    """
    Gamma distribution 
    """
    def __init__(self, alpha, beta):
        self.alpha = alpha
        self.beta = beta
        self.dist = gamma(a=alpha,scale=1/beta)

    def pdf(self, x):
        return self.dist.pdf(x)

    def logpdf(self, x):
        return self.dist.logpdf(x)
    
    def sample(self, n):
        return self.dist.rvs(n)
```

```python
def model_logposterior(mu, sigma, y):
    """
    Model log-posterior
    """
    # log-probability of mu under prior.
    normal_prior = Normal(0, 1)
    mu_log_prob = normal_prior.logpdf(mu)

    # log-probability of sigma under prior.
    sigma_prior = Gamma(11,1)
    sigma_log_prob = sigma_prior.logpdf(sigma)

    # log-likelihood given priors and data
    likelihood = Normal(mu, sigma)
    likelihood_log_prob = likelihood.logpdf(y).sum()

    # Joint log-likelihood
    return mu_log_prob + sigma_log_prob + likelihood_log_prob
```

# Metropolis-Hastings Sampling

```python
def metropolis_hastings(y,model,params,iterations):

    """
    Metropolis-Hastings Sampling
    """
    # Initial parameters
    mu_old=params[0]
    sigma_old=params[1]
    
    # Keep a history of the parameter values and ratio.
    mu_history = []
    sigma_history = [] 
    ratio_history=[]
    joint_posterior=[]
    accepted=[]
    
    for i in range(iterations):
        mu_history.append(mu_old)
        sigma_history.append(sigma_old)
        
        mu_new=np.random.normal(mu_old,params[1])
        sigma_new=np.random.normal(sigma_old, params[1])
        
        llik_old = model(mu_old, sigma_old, y) 
        llik_new = model(mu_new, sigma_new, y)  
        
        if llik_new>llik_old or (np.log(np.random.uniform(0,1)) < (llik_new-llik_old)):            
            mu_old=mu_new
            sigma_old=sigma_new
            accepted.append(1)
        else:
            accepted.append(0)
            
        ratio_history.append(llik_new-llik_old)
        joint_posterior.append(llik_old)
        
    mu_history=np.array(mu_history)
    sigma_history=np.array(sigma_history)
    ratio_history=np.array(ratio_history)
    accepted=np.array(accepted)
    joint_posterior=np.array(joint_posterior)
    print('Acceptance rate',round(np.sum(accepted)/iterations,2))
    return mu_history,sigma_history,joint_posterior,ratio_history,accepted
```

```python
niter=10**4
mu_history,sigma_history,joint_posterior,ratio,accepted=metropolis_hastings(yobs,model_logposterior,
                                                               [np.mean(yobs),0.07*np.std(yobs)],
                                                               iterations=niter)
```

# Pymc3

```python
import pymc3 as pm
```

```python
with pm.Model() as model: 
    # Define priors
    mu = pm.Normal('mu', mu=0, sigma=1)
    sigma = pm.Gamma('sigma',alpha=11,beta=1)
    
    # Define likelihood
    y = pm.Normal('y', mu=mu,sigma=sigma, observed=yobs)
```

```python
# Inference!
with model:
    trace = pm.sample(2000, tune=1000, chains=2,target_accept=0.95)
```

```python
f, axs = plt.subplots(nrows=2, ncols=2, figsize=stdfigsize(nx=2, ny=2, scale=1))

ax=axs.flatten()[0]
ax.plot(trace['mu'][-1000:])
ax.set_xlabel("Iteration")
ax.set_ylabel(r"$\mu$")

ax=axs.flatten()[1]
sns.histplot(trace['mu'][-1000:],ax=ax,kde=True)
ax.set_xlabel(r"$\mu$")
ax.set_ylabel(r"PDF")

ax=axs.flatten()[2]
ax.plot(trace['sigma'][-1000:])
ax.set_xlabel("Iteration")
ax.set_ylabel(r"$\sigma$")

ax=axs.flatten()[3]
sns.histplot(trace['sigma'][-1000:],ax=ax,kde=True)
ax.set_xlabel(r"$\sigma$")
ax.set_ylabel(r"PDF")
f.tight_layout()
```
